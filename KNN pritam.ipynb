```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from collections import Counter
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,0:8]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))
```

```python
def vote(neighbours): 
    class_counter = Counter(neighbours) 
    return class_counter.most_common(1)[0][0] 
```

```python
def knn(x_train, y_train, test_point, k): 
    distances = [] 
    for i in range(len(x_train)): 
        distance = euclidean_distance(test_point, x_train.iloc[i]) 
        distances.append((distance, y_train.iloc[i])) 
    distances = sorted(distances) 
    neighbours = np.asarray(distances[:k]) 
    label = vote(neighbours[:, 1]) 
    return label 
```

```python
predictions = [] 
for i in range(len(X_test)): 
    predict = knn(X_train, y_train, X_test.iloc[i], 5) 
    predictions.append(predict) 
```

```python
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy of KNN: {accuracy}")
```
